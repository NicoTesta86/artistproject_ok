```python
import pandas as pd

# Load the Career Performance data
career_performance_path = 'Nora En Pure-Analytics-Career Performance.csv'
career_performance_df = pd.read_csv(career_performance_path)

# Display the first few rows of the dataframe to understand its structure
career_performance_df.head()
```

The Career Performance data for Nora En Pure includes the following columns:

    Date: The date of the recorded data.
    Viberate performance points: Points indicating Nora En Pure's performance in the music industry, as assessed by Viberate.
    Viberate rank: Nora En Pure's rank on Viberate for each date. It seems like the first entry is missing this value, which might indicate an initial recording or an error.

To analyze Nora En Pure's career performance, we can:

    Plot the performance points over time to visualize trends and identify peak periods.
    Look at the ranking trend to see how her position in the industry has changed.

```python
import matplotlib.pyplot as plt

# Convert 'Date' to datetime format
career_performance_df['Date'] = pd.to_datetime(career_performance_df['Date'])

# Plotting
fig, ax1 = plt.subplots(figsize=(14, 7))

color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Viberate performance points', color=color)
ax1.plot(career_performance_df['Date'], career_performance_df['Viberate performance points'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Viberate rank', color=color)  # we already handled the x-label with ax1
ax2.plot(career_performance_df['Date'], career_performance_df['Viberate rank'], color=color)
ax2.tick_params(axis='y', labelcolor=color)
ax2.invert_yaxis()  # Invert y axis to show higher ranks at the top

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('Nora En Pure Career Performance and Rank Over Time')
plt.show()
```

The plot visualizes Nora En Pure's career performance in terms of Viberate performance points (red) and Viberate rank (blue) over time. A few observations:

    Performance Points: We see fluctuations in the performance points, indicating periods of increased activity, success, or visibility in the industry. These could correspond to new releases, successful collaborations, or significant live performances.

    Rank: The rank, when lower in value, indicates a better position. It's inversely plotted to make this more intuitive, with improvements in rank shown by a downward trend on the graph. There are fluctuations here as well, suggesting changes in Nora En Pure's standing relative to other artists in the industry.

```python
# Load the Fanbase Evolution Through Time data
fanbase_evolution_path = 'Nora En Pure-Analytics-Fanbase Evolution Through Time.csv'
fanbase_evolution_df = pd.read_csv(fanbase_evolution_path)

# Display the first few rows of the dataframe to understand its structure
fanbase_evolution_df.head()
```


Moving on to the Fanbase Analysis, we'll explore various aspects of Nora En Pure's fanbase, including its evolution over time, daily growth and engagement trends, and the overall distribution and growth of the fanbase. This comprehensive view will help us understand how her audience has developed and engaged with her over time.

Let's start with the Fanbase Evolution Through Time using the "Nora En Pure-Analytics-Fanbase Evolution Through Time.csv" file to observe how Nora En Pure's fanbase has changed.

The "Fanbase Evolution Through Time" dataset includes the following key metrics recorded over various dates:

    Total and new followers/subscribers/fans on Spotify, YouTube, Instagram, Facebook, Deezer, X (possibly another platform), and Soundcloud. TikTok data seems to be unavailable (NaN values).
    The data provides a comprehensive view of Nora En Pure's presence across major social media and streaming platforms, tracking both the total number of followers and the new followers gained over each recorded period.

To analyze Nora En Pure's fanbase evolution, we can:

    Visualize the total followers across all platforms over time to understand overall growth trends.
    Compare the growth rates between platforms to see where Nora En Pure is gaining the most traction.
    Identify any significant spikes in new followers which might indicate successful releases, collaborations, or marketing efforts.

```python
# Convert 'Date' to datetime format for plotting
fanbase_evolution_df['Date'] = pd.to_datetime(fanbase_evolution_df['Date'])

# Plotting total followers for each platform over time
plt.figure(figsize=(14, 8))

# Spotify
plt.plot(fanbase_evolution_df['Date'], fanbase_evolution_df['Spotify followers total'], label='Spotify', marker='o')

# Youtube
plt.plot(fanbase_evolution_df['Date'], fanbase_evolution_df['Youtube subscribers total'], label='Youtube', marker='o')

# Instagram
plt.plot(fanbase_evolution_df['Date'], fanbase_evolution_df['Instagram followers total'], label='Instagram', marker='o')

# Facebook
plt.plot(fanbase_evolution_df['Date'], fanbase_evolution_df['Facebook followers total'], label='Facebook', marker='o')

# Deezer
plt.plot(fanbase_evolution_df['Date'], fanbase_evolution_df['Deezer fans total'], label='Deezer', marker='o')

# X (Unknown Platform)
plt.plot(fanbase_evolution_df['Date'], fanbase_evolution_df['X followers total'], label='X (Unknown Platform)', marker='o')

# Soundcloud
plt.plot(fanbase_evolution_df['Date'], fanbase_evolution_df['Soundcloud followers total'], label='Soundcloud', marker='o')

plt.title('Nora En Pure Fanbase Evolution Across Platforms')
plt.xlabel('Date')
plt.ylabel('Total Followers')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout to make room for the rotated x-axis labels

plt.show()
```


The plot illustrates the evolution of Nora En Pure's fanbase across various platforms over time. Key observations include:

    Steady Growth: There's a consistent upward trend in followers across all platforms, indicating a growing interest and engagement with Nora En Pure's music and content.

    Platform Variation: The growth patterns vary between platforms, with some platforms showing more rapid increases. This could reflect the varying nature of content consumption across these platforms and Nora En Pure's focus or success in engaging with her audience on specific ones.

    Major Platforms: Spotify and Instagram show significant total followers, highlighting their importance in Nora En Pure's social media and streaming strategy. YouTube and Facebook also play crucial roles in her fanbase engagement.

This growth trend is a positive sign, reflecting ongoing engagement and an expanding audience. To delve deeper, we could analyze new followers' spikes to correlate them with specific releases or events, providing insights into what drives fanbase growth for Nora En Pure.

Let's move on to analyzing the Daily Growth and Engagement of Nora En Pure's fanbase. We'll use the "Nora En Pure-Analytics-Fanbase Daily Growth and Engagement.csv" file for this analysis. I'll start by loading the data and providing a summary of its contents.

The "Fanbase Daily Growth and Engagement" dataset includes detailed daily metrics for Nora En Pure, covering:

    Spotify: Followers, followers change, monthly listeners, and monthly listeners change.
    YouTube: Subscribers and subscribers change.
    TikTok: Followers, followers change, likes, and views (though it seems we have missing data for TikTok metrics).
    Instagram: Followers, followers change, and likes.
    SoundCloud: Followers, followers change, and plays.

This dataset offers a granular look at how Nora En Pure's fanbase reacts to new content, social media activity, and other factors on a day-to-day basis. For our analysis, we can:

    Visualize the daily change in followers across platforms to identify any spikes or trends.
    Look at the engagement metrics like Spotify monthly listeners change, Instagram likes, and SoundCloud plays to correlate engagement with follower growth

```python
# Convert 'Date' to datetime format for plotting
daily_growth_engagement_df['Date'] = pd.to_datetime(daily_growth_engagement_df['Date'])

# Plotting daily change in followers for selected platforms
plt.figure(figsize=(14, 8))

# Spotify followers change
plt.plot(daily_growth_engagement_df['Date'], daily_growth_engagement_df['Spotify followers change'], label='Spotify Followers Change', marker='o')

# Youtube subscribers change
plt.plot(daily_growth_engagement_df['Date'], daily_growth_engagement_df['Youtube subscribers change'], label='Youtube Subscribers Change', marker='o')

# Instagram followers change
plt.plot(daily_growth_engagement_df['Date'], daily_growth_engagement_df['Instagram followers change'], label='Instagram Followers Change', marker='o')

# SoundCloud followers change
plt.plot(daily_growth_engagement_df['Date'], daily_growth_engagement_df['SoundCloud followers change'], label='SoundCloud Followers Change', marker='o')

plt.title('Daily Change in Followers/Subscribers Across Platforms')
plt.xlabel('Date')
plt.ylabel('Change in Followers/Subscribers')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout to make room for the rotated x-axis labels

plt.show()
```


The plot showcases the daily change in followers/subscribers across several platforms for Nora En Pure. Here are some observations:

    Spotify: Shows a consistent daily increase in followers, indicating steady growth on this platform. This is a good sign of ongoing interest in Nora En Pure's music on Spotify.

    YouTube: There's a significant spike in subscribers on one of the days, possibly indicating a new video release or a viral content piece that attracted more subscribers.

    Instagram: Like Spotify, Instagram shows a consistent increase in followers, with some days experiencing higher growth. This could be related to specific posts or stories that resonated well with the audience.

    SoundCloud: The platform shows a steady, albeit smaller, daily increase in followers. The consistent growth across platforms highlights Nora En Pure's broad appeal and the effectiveness of her content strategy.
