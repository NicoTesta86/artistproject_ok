```python
import pandas as pd


data = pd.read_csv('Giordano-Analytics-Fanbase Total Distribution(1) copia.csv')

data.head()
```

###### analisi della distribuzione tramite grafico ad isto

```python
import matplotlib.pyplot as plt

# Create a figure and a set of subplots for both total numbers and percentages
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar plot for the total followers/fans
ax1.bar(data['Channel'], data['Total'], color='skyblue', label='Total Followers/Fans')
ax1.set_xlabel('Channel')
ax1.set_ylabel('Total Followers/Fans', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')
ax1.set_title('Distribution of Fan Base Across Channels')

# Instantiate a second axes that shares the same x-axis
ax2 = ax1.twinx()  
ax2.plot(data['Channel'], data['Percentage'], color='darkgreen', marker='o', label='Percentage of Total')
ax2.set_ylabel('Percentage (%)', color='darkgreen')  
ax2.tick_params(axis='y', labelcolor='darkgreen')

# Show legend
fig.tight_layout()
plt.show()
```

Considerazione di grafico: abbiamo tutto focalizzato su un pubblico si instagram e facebook,  e viene poco nulla valorizzato il mercato degli streaming come spotify. Soundcloud è buono ma non sviluppato troppo

```python
# Prima di procedere con la creazione di un grafico, apriamo e esaminiamo il nuovo file caricato per capire i suoi dati.
file_path_country = 'Giordano-Audience By Country.csv'
data_country = pd.read_csv(file_path_country)

# Visualizziamo le prime righe per comprendere la struttura del file
data_country.head()
```

```python
# Creiamo un grafico a barre per visualizzare la percentuale dell'audience totale su Viberate e gli ascoltatori mensili su Spotify
# per i primi cinque paesi.

# Selezioniamo solo i primi cinque paesi per semplicità
top_countries = data_country.head()

# Creazione del grafico
fig, ax1 = plt.subplots(figsize=(10, 6))

# Asse principale: Viberate overall audience (%)
ax1.bar(top_countries['Country'], top_countries['Viberate overall audience (%)'], color='lightblue', label='Viberate Overall Audience (%)')
ax1.set_xlabel('Country')
ax1.set_ylabel('Viberate Overall Audience (%)', color='lightblue')
ax1.tick_params(axis='y', labelcolor='lightblue')
ax1.set_title('Audience Distribution by Country')

# Asse secondario: Spotify monthly listeners
ax2 = ax1.twinx()
ax2.plot(top_countries['Country'], top_countries['Spotify monthly listeners'], color='darkgreen', marker='o', linestyle='-', label='Spotify Monthly Listeners')
ax2.set_ylabel('Spotify Monthly Listeners', color='darkgreen')
ax2.tick_params(axis='y', labelcolor='darkgreen')

fig.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load the CSV file to see the first few rows and understand its structure
data_path = 'Giordano-Analytics-Fanbase Daily Growth and Engagement(1).csv'
data = pd.read_csv(data_path)

# Display the first few rows of the dataframe
data.head()
```

#### Trend di crescita dei follower

```python
import matplotlib.pyplot as plt

# Convert 'Date' column to datetime format for plotting
data['Date'] = pd.to_datetime(data['Date'])

# Plotting the growth trend of followers/subscribers for available platforms
plt.figure(figsize=(14, 10))

# Spotify Followers
plt.plot(data['Date'], data['Spotify followers'], label='Spotify Followers')

# YouTube Subscribers
plt.plot(data['Date'], data['Youtube subscribers'], label='YouTube Subscribers')

# TikTok Followers
plt.plot(data['Date'], data['TikTok followers'], label='TikTok Followers')

# Instagram Followers
plt.plot(data['Date'], data['Instagram followers'], label='Instagram Followers')

# SoundCloud Followers
plt.plot(data['Date'], data['SoundCloud followers'], label='SoundCloud Followers')

plt.title('Growth Trend of Followers/Subscribers Across Platforms')
plt.xlabel('Date')
plt.ylabel('Number of Followers/Subscribers')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()

# Display the plot
plt.show()
```
