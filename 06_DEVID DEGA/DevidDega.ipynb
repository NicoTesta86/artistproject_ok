```python
import pandas as pd

# Load the data from "Fanbase Evolution Through Time"
fanbase_evolution_path = 'DEVID DEGA-Analytics-Fanbase Evolution Through Time.csv'
fanbase_evolution_df = pd.read_csv(fanbase_evolution_path)

# Display the first few rows to understand the structure of the data
fanbase_evolution_df.head()
```

To begin, we need to first examine one of the files to better understand the structure of the available data. Let's start by looking into the "Fanbase Evolution Through Time" data as our first step.

The "Fanbase Evolution Through Time" data provides a detailed overview of fanbase growth across multiple platforms, including Spotify, YouTube, Instagram, Facebook, Deezer, X (formerly known as Twitter), and SoundCloud. Each row represents a weekly observation that includes both the total number of followers/fans and the number of new followers/fans gained in that week.

We can proceed in several ways for our analysis. One approach would be to visualize the total fanbase growth over time for each platform, identifying trends and patterns. Another could be to compare the growth of new followers across different platforms over the same time period, to see which channels have been most effective in growing the fanbase.

Given that we've identified an initial area of interest, let's start by creating a line chart to visualize the total fanbase growth over time across different platforms. This will give us an immediate view of which platforms have performed the best and how the fanbase has grown over time

```python
import matplotlib.pyplot as plt

# Convert the 'Date' column to datetime format
fanbase_evolution_df['Date'] = pd.to_datetime(fanbase_evolution_df['Date'])

# Plotting the total followers for each platform over time
plt.figure(figsize=(14, 8))

# Spotify
plt.plot(fanbase_evolution_df['Date'], fanbase_evolution_df['Spotify followers total'], label='Spotify', marker='o')

# YouTube
plt.plot(fanbase_evolution_df['Date'], fanbase_evolution_df['Youtube subscribers total'], label='YouTube', marker='o')

# Instagram
plt.plot(fanbase_evolution_df['Date'], fanbase_evolution_df['Instagram followers total'], label='Instagram', marker='o')

# Facebook
plt.plot(fanbase_evolution_df['Date'], fanbase_evolution_df['Facebook followers total'], label='Facebook', marker='o')

# Deezer
plt.plot(fanbase_evolution_df['Date'], fanbase_evolution_df['Deezer fans total'], label='Deezer', marker='o')

# X (formerly Twitter)
plt.plot(fanbase_evolution_df['Date'], fanbase_evolution_df['X followers total'], label='X', marker='o')

# SoundCloud
plt.plot(fanbase_evolution_df['Date'], fanbase_evolution_df['Soundcloud followers total'], label='SoundCloud', marker='o')

plt.title('Fanbase Evolution Over Time Across Platforms')
plt.xlabel('Date')
plt.ylabel('Total Followers/Fans')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The chart above visualizes the evolution of the fanbase over time across multiple platforms, including Spotify, YouTube, Instagram, Facebook, Deezer, X (formerly known as Twitter), and SoundCloud. Each line represents the total number of followers or fans on a specific platform, showcasing how the fanbase has grown over the observed period.

From this visualization, we can begin to identify trends, such as which platforms have seen the most significant growth, periods of rapid increases in followers, and potentially, platforms that have not performed as well.

```python
# Load the data from "Engagement"
engagement_path = 'DEVID DEGA-Analytics-Engagement.csv'
engagement_df = pd.read_csv(engagement_path)

# Display the first few rows to understand the structure of the data
engagement_df.head()
```

The next step in our analysis focuses on Engagement Analysis. We aim to understand the relationship between daily engagement and fanbase growth, identifying which types of content or activities drive the highest engagement levels. To do this, we'll analyze the data from the "Engagement" and "Fanbase Daily Growth and Engagement" files.

First, we need to examine the structure of the engagement data to determine how best to visualize and analyze these relationships. Let's start by looking into the "Engagement" data file to understand its format and the type of insights we might be able to extract from it.

The "Engagement" data file provides a weekly overview of engagement metrics across various platforms, including Spotify streams, YouTube views, Instagram post likes, Shazams, airplay spins, X likes (formerly Twitter), SoundCloud plays, and TikTok likes. Each row corresponds to a week, showing how different types of content and activities have engaged the audience.

To analyze engagement, we can:

- Visualize the weekly engagement metrics for each platform to observe trends and identify which platforms have the highest engagement levels.
- Compare the growth in engagement over time with the fanbase growth data to see if there's a correlation between an increase in followers and engagement levels.

Given the available data, creating a set of line charts for each engagement metric over time could provide us with insights into which content types or platforms generate the most engagement. This approach will help us identify high-performing content and periods where engagement spiked significantly.

```python
# Convert the 'Date' column to datetime format for engagement data
engagement_df['Date'] = pd.to_datetime(engagement_df['Date'])

# Plotting the engagement metrics over time
plt.figure(figsize=(16, 10))

# Spotify streams
plt.subplot(3, 3, 1)
plt.plot(engagement_df['Date'], engagement_df['Spotify streams'], label='Spotify Streams', marker='o')
plt.title('Spotify Streams')
plt.xticks(rotation=45)

# Youtube views
plt.subplot(3, 3, 2)
plt.plot(engagement_df['Date'], engagement_df['Youtube views'], label='YouTube Views', marker='o')
plt.title('YouTube Views')
plt.xticks(rotation=45)

# Instagram post likes
plt.subplot(3, 3, 3)
plt.plot(engagement_df['Date'], engagement_df['Instagram post likes'], label='Instagram Likes', marker='o')
plt.title('Instagram Post Likes')
plt.xticks(rotation=45)

# Shazams
plt.subplot(3, 3, 4)
plt.plot(engagement_df['Date'], engagement_df['Shazams'], label='Shazams', marker='o')
plt.title('Shazams')
plt.xticks(rotation=45)

# Airplay spins
plt.subplot(3, 3, 5)
plt.plot(engagement_df['Date'], engagement_df['Airplay spins'], label='Airplay Spins', marker='o')
plt.title('Airplay Spins')
plt.xticks(rotation=45)

# X likes
plt.subplot(3, 3, 6)
plt.plot(engagement_df['Date'], engagement_df['X likes'], label='X Likes', marker='o')
plt.title('X Likes')
plt.xticks(rotation=45)

# Soundcloud plays
plt.subplot(3, 3, 7)
plt.plot(engagement_df['Date'], engagement_df['Soundcloud plays'], label='SoundCloud Plays', marker='o')
plt.title('SoundCloud Plays')
plt.xticks(rotation=45)

# Since TikTok likes contain NaN values, they're not plotted here.

plt.tight_layout()
plt.show()
```


The set of line charts above visualizes the weekly engagement metrics across various platforms, including Spotify streams, YouTube views, Instagram post likes, Shazams, airplay spins, X likes, and SoundCloud plays. These visualizations help us observe the trends in engagement over time and identify which platforms and types of content have generated the highest levels of engagement.

Some key observations might include:

- Trends and Patterns: Identifying any noticeable trends in engagement, such as consistent growth or periodic spikes that may align with specific events or content releases.
- Platform Performance: Seeing which platforms consistently have higher engagement metrics, which can inform future content and marketing strategies.
- Engagement Peaks: Noting any periods where there was a significant spike in engagement, which could be correlated with specific promotional activities or content releases.

```python
# Load the data from "Top Tracks on Music Channels"
top_tracks_path = 'DEVID DEGA-Analytics-Top Tracks on Music Channels.csv'
top_tracks_df = pd.read_csv(top_tracks_path)

# Display the first few rows to understand the structure of the data
top_tracks_df.head()
```

For the final part of our analysis, we'll focus on Career Performance and Top Tracks Analysis. This involves examining the overall career performance and identifying the most popular tracks across different music channels. Using data from "Career Performance" and "Top Tracks on Music Channels", we can:

- Create a bar chart to showcase the most listened tracks on various music channels.
- Analyze the correlation between track popularity and periods of fanbase growth or spikes in engagement.

First, let's take a look at the "Top Tracks on Music Channels" data to understand its structure and determine how we can best visualize the popularity of tracks across different platforms. This will help us identify the tracks that have significantly contributed to the artist's popularity and fanbase growth.

The "Top Tracks on Music Channels" data provides information on the most popular tracks across different channels, such as Shazam in this case, including track names, artists, release dates, and engagement metrics over different time periods (week, month, and total).

To visualize this data effectively and analyze the popularity of tracks, we can:

- Aggregate the total engagement for each track across all channels (if data for more channels were available) to identify the most popular tracks.
- Create a bar chart showcasing the top tracks based on their total engagement, which will help us see which tracks have resonated the most with the audience.

Given the structure of this data, focusing on total engagement seems most appropriate for our analysis. Let's proceed by visualizing the top tracks based on their "Engagement Total" to understand their impact on the artist's career and fanbase growth

```python
# Since the dataset might contain NaN values for engagement, we'll fill them with 0 for visualization purposes
top_tracks_df['Engagement Total'].fillna(0, inplace=True)

# Sorting the tracks by total engagement and selecting the top 10 for visualization
top_tracks_sorted = top_tracks_df.sort_values(by='Engagement Total', ascending=False).head(10)

# Creating a bar chart
plt.figure(figsize=(12, 8))
plt.barh(top_tracks_sorted['Track'], top_tracks_sorted['Engagement Total'], color='skyblue')
plt.xlabel('Total Engagement')
plt.ylabel('Tracks')
plt.title('Top 10 Tracks by Total Engagement on Music Channels')
plt.tight_layout()
plt.show()

```

The bar chart above showcases the top 10 tracks by total engagement on music channels, highlighting those that have garnered the most attention and engagement from the audience. This visualization helps identify which tracks have been pivotal in building the artist's popularity and fanbase.

By analyzing the engagement levels of these top tracks, we can infer:

- Key Contributions: Which tracks have significantly contributed to the artist's career based on audience engagement.
- Potential Trends: If certain styles or collaborations are more popular than others, suggesting directions for future projects.
- Correlation with Fanbase Growth: How the release and popularity of these tracks may correlate with observed spikes in fanbase growth or engagement across platforms.
