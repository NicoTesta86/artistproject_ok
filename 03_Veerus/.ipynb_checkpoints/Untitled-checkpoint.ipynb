```python
import pandas as pd


career_performance_path = 'Veerus-Analytics-Career Performance(1).csv'
career_performance_df = pd.read_csv(career_performance_path)

career_performance_df
```

```python
import matplotlib.pyplot as plt


career_performance_df['Date'] = pd.to_datetime(career_performance_df['Date'])

plt.figure(figsize=(14, 7))
plt.plot(career_performance_df['Date'], career_performance_df['Viberate performance points'], marker='o', linestyle='-', color='blue')
plt.title('Trend pt Performance in time')
plt.xlabel('date')
plt.ylabel('Pt Performance')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
career_performance_filtered = career_performance_df.dropna(subset=['Viberate rank'])

plt.figure(figsize=(14, 7))
plt.plot(career_performance_filtered['Date'], career_performance_filtered['Viberate rank'], marker='o', linestyle='-', color='red')
plt.title('Veerus Ranking Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Viberate Rank')
plt.gca().invert_yaxis()  # Inverting y-axis as a lower rank number is better
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

The graph depicts the trend of Veerus's ranking over time. Note that the y-axis is inverted, meaning a lower rank (closer to 1) is considered better. The plot shows how the ranking has changed, possibly reflecting the artist's or entity's relative standing within their industry or genre.

Observing the ranking trend can provide insights into periods of significant improvement or decline in Veerus's industry standing. These changes could be linked to specific career events, releases, or external factors affecting the broader industry landscape.



```python
fig, ax1 = plt.subplots(figsize=(14, 7))

color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Viberate Performance Points', color=color)
ax1.plot(career_performance_filtered['Date'], career_performance_filtered['Viberate performance points'], color=color, label='Performance Points')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:blue'
ax2.set_ylabel('Viberate Rank', color=color)  # we already handled the x-label with ax1
ax2.plot(career_performance_filtered['Date'], career_performance_filtered['Viberate rank'], color=color, label='Rank')
ax2.tick_params(axis='y', labelcolor=color)
ax2.invert_yaxis()  # Inverting y-axis to keep the rank's orientation consistent

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('Performance Points and Viberate Rank Over Time')
plt.show()
```

This different visualization combines the trends of both performance points and Viberate rank over time, using a dual-axis line chart. The red line represents the Viberate performance points, and the blue line represents the Viberate rank. The y-axis on the left corresponds to the performance points, and the y-axis on the right, which is inverted, corresponds to the rank.

By comparing these two metrics on the same timeline, we can more easily see how changes in performance points are reflected in the Viberate rank. This visualization emphasizes the relationship between performance improvements and ranking adjustments, providing a clear view of how they have evolved together over time.

## Engagement

The "Engagement" file provides a detailed look at various metrics related to Veerus's audience engagement across different platforms and media for specific dates. Here are the columns included in the dataset:

    Date: The date of the engagement data record.
    Spotify streams: The number of streams on Spotify.
    Youtube views: The number of views on Youtube.
    Instagram post likes: The number of likes on Instagram posts.
    Shazams: The number of times Veerus's songs were identified on Shazam.
    Airplay spins: The number of times Veerus's songs were played on the radio.
    X likes: This column's context isn't clear without additional information, but it could be another social media engagement metric.
    Soundcloud plays: The number of plays on Soundcloud.
    Tiktok likes: The number of likes on Tiktok.

```python

engagement_path = 'Veerus-Analytics-Engagement.csv'
engagement_df = pd.read_csv(engagement_path)


engagement_df.head()
```

```python
# Convertire la colonna "Date" in datetime per una migliore gestione dei dati temporali
engagement_df['Date'] = pd.to_datetime(engagement_df['Date'])

# Grafico delle tendenze di engagement nel tempo per ogni piattaforma
plt.figure(figsize=(16, 10))

# Spotify streams
plt.plot(engagement_df['Date'], engagement_df['Spotify streams'], label='Spotify Streams', marker='o', linestyle='-')
# Youtube views
plt.plot(engagement_df['Date'], engagement_df['Youtube views'], label='Youtube Views', marker='o', linestyle='-')
# Instagram post likes
plt.plot(engagement_df['Date'], engagement_df['Instagram post likes'], label='Instagram Post Likes', marker='o', linestyle='-')
# Shazams
plt.plot(engagement_df['Date'], engagement_df['Shazams'], label='Shazams', marker='o', linestyle='-')
# Airplay spins
plt.plot(engagement_df['Date'], engagement_df['Airplay spins'], label='Airplay Spins', marker='o', linestyle='-')
# X likes
plt.plot(engagement_df['Date'], engagement_df['X likes'], label='X Likes', marker='o', linestyle='-')
# Soundcloud plays
plt.plot(engagement_df['Date'], engagement_df['Soundcloud plays'], label='Soundcloud Plays', marker='o', linestyle='-')
# Tiktok likes
plt.plot(engagement_df['Date'], engagement_df['Tiktok likes'], label='Tiktok Likes', marker='o', linestyle='-')

plt.title('Engagement Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Engagement Metrics')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

The chart illustrates the engagement trends over time across various platforms for Veerus. Each line represents a different engagement metric, including Spotify streams, Youtube views, Instagram post likes, Shazams, Airplay spins, X likes, Soundcloud plays, and Tiktok likes.

Key observations from the chart might include:

    Variability: Engagement metrics vary significantly across platforms, with some showing more pronounced fluctuations than others.
    Platform Popularity: Some platforms, like Spotify and Youtube, might show consistently higher engagement, suggesting they are more popular or effective channels for Veerus.
    Special Events: Spikes or dips in the engagement metrics might correlate with special events, releases, or promotional activities.
    Emerging Trends: New or growing platforms might show increasing engagement over time, indicating shifting audience behaviors or successful outreach efforts.

This overview provides a broad picture of how Veerus's audience interacts with their content across different media. Further analysis could delve into the causes behind trends, the impact of specific campaigns, or comparisons between engagement types to derive actionable insights.

## Audience Map

The "Audience Map" file provides detailed data on Veerus's audience distribution across different countries and platforms. Here's a summary of the key columns in the dataset:

    Country: The country of the audience.
    Viberate overall audience (%): The percentage of the overall audience located in each country.
    Spotify monthly listeners: The number of monthly listeners on Spotify from each country.
    Spotify monthly listeners (%): The percentage of Spotify monthly listeners from each country relative to the total.
    Spotify monthly listeners growth: Growth in the number of Spotify monthly listeners (the dataset doesn't show this data for now).
    Youtube monthly views: The number of monthly views on Youtube from each country.
    Youtube monthly views (%): The percentage of Youtube monthly views from each country relative to the total.
    Youtube monthly views growth: Growth in the number of Youtube monthly views (the dataset doesn't show this data for now).
    Youtube subscribers: The number of Youtube subscribers from each country.
    Youtube subscribers (%): The percentage of Youtube subscribers from each country relative to the total.
    Instagram followers: The number of Instagram followers from each country.
    Instagram followers (%): The percentage of Instagram followers from each country relative to the total.
    Tiktok followers: The number of Tiktok followers from each country.
    Tiktok followers (%): The percentage of Tiktok followers from each country relative to the total.

```python

audience_map_path = 'Veerus-Analytics-Audience-Map.csv'
audience_map_df = pd.read_csv(audience_map_path)

audience_map_df.head()
```

```python

audience_map_df['Viberate overall audience (%)'] = audience_map_df['Viberate overall audience (%)'].str.rstrip('%').astype('float') / 100.0


plt.figure(figsize=(20, 20))
audience_map_df.plot(kind='bar', x='Country', y='Viberate overall audience (%)', color='darkcyan')
plt.title('Geographical Distribution of Audience')
plt.xlabel('Country')
plt.ylabel('Percentage of Overall Audience')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

The bar chart illustrates the geographical distribution of Veerus's audience, showing the percentage of the overall audience by country. This visualization helps identify where Veerus's audience is most concentrated geographically.

Countries with higher percentages represent larger portions of Veerus's total audience, indicating key markets or regions where Veerus has a strong presence.
