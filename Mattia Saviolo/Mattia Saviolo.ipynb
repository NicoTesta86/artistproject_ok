## MATTIA SAVIOLO 

```python
import pandas as pd

# Load the CSV file to analyze Mattia Saviolo's career performance
data_path = 'Mattia Saviolo-Analytics-Career Performance.csv'
df = pd.read_csv(data_path)

# Display the first few rows of the dataframe to understand its structure
df.head()
```

## PERFORMANCE

```python
import matplotlib.pyplot as plt

# Convert 'Date' to datetime format for better plotting
df['Date'] = pd.to_datetime(df['Date'])

# Plotting the Viberate performance points over time
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Viberate performance points'], marker='o', linestyle='-', color='blue')
plt.title('Mattia Saviolo Viberate Performance Points Over Time')
plt.xlabel('Date')
plt.ylabel('Viberate Performance Points')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Save and show the plot
plt_path = 'mattia_saviolo_performance_plot.png'
plt.savefig(plt_path)
plt.show()

plt_path
```

```python
# Identifying points of increase in Viberate performance points
df['Increase'] = df['Viberate performance points'].diff() > 0

# Plotting with emphasis on points of increase
plt.figure(figsize=(12, 7))

# Plot all data points
plt.plot(df['Date'], df['Viberate performance points'], label='Performance Points', color='lightgray', marker='o', linestyle='-')

# Highlight points of increase
plt.scatter(df[df['Increase']]['Date'], df[df['Increase']]['Viberate performance points'], color='red', label='Points of Increase', zorder=5)

plt.title('Mattia Saviolo Viberate Performance Points with Emphasis on Increases')
plt.xlabel('Date')
plt.ylabel('Viberate Performance Points')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Save and show the modified plot
plt_path_increases = 'mattia_saviolo_performance_increase_plot.png'
plt.savefig(plt_path_increases)
plt.show()

plt_path_increases
```

```python
# Identifying points of decrease in Viberate performance points
df['Decrease'] = df['Viberate performance points'].diff() < 0

# Plotting with emphasis on points of decrease
plt.figure(figsize=(12, 7))

# Plot all data points
plt.plot(df['Date'], df['Viberate performance points'], label='Performance Points', color='lightgray', marker='o', linestyle='-')

# Highlight points of decrease
plt.scatter(df[df['Decrease']]['Date'], df[df['Decrease']]['Viberate performance points'], color='green', label='Points of Decrease', zorder=5)

plt.title('Mattia Saviolo Viberate Performance Points with Emphasis on Decreases')
plt.xlabel('Date')
plt.ylabel('Viberate Performance Points')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Save and show the modified plot
plt_path_decreases = 'mattia_saviolo_performance_decrease_plot.png'
plt.savefig(plt_path_decreases)
plt.show()

plt_path_decreases
```

```python
# Counting the number of increase and decrease points
increase_count = df['Increase'].sum()
decrease_count = df['Decrease'].sum()

increase_count, decrease_count
```

```python
# Load the CSV file to analyze daily growth and engagement for SoundCloud, Spotify, and Instagram
growth_path = 'Mattia Saviolo-Analytics-Fanbase Daily Growth and Engagement(1).csv'
growth_df = pd.read_csv(growth_path)

# Display the first few rows of the dataframe to understand its structure
growth_df.head()
```

## SPOTIFY

```python
# Convert 'Date' to datetime for better plotting
growth_df['Date'] = pd.to_datetime(growth_df['Date'])

# Plotting Spotify followers and monthly listeners over time
fig, ax1 = plt.subplots(figsize=(12, 7))

color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Spotify Followers', color=color)
ax1.plot(growth_df['Date'], growth_df['Spotify followers'], color=color, label='Spotify Followers')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:blue'
ax2.set_ylabel('Spotify Monthly Listeners', color=color)  # we already handled the x-label with ax1
ax2.plot(growth_df['Date'], growth_df['Spotify monthly listeners'], color=color, label='Spotify Monthly Listeners')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('Spotify Followers and Monthly Listeners Over Time')
plt.show()
```

## INSTAGRAM

```python
# Plotting Instagram followers and likes over time
fig, ax1 = plt.subplots(figsize=(12, 7))

color = 'tab:green'
ax1.set_xlabel('Date')
ax1.set_ylabel('Instagram Followers', color=color)
ax1.plot(growth_df['Date'], growth_df['Instagram followers'], color=color, label='Instagram Followers')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:orange'
ax2.set_ylabel('Instagram Likes', color=color)  # we already handled the x-label with ax1
ax2.plot(growth_df['Date'], growth_df['Instagram likes'], color=color, label='Instagram Likes', linestyle='--')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('Instagram Followers and Likes Over Time')
plt.show()
```

## SOUNDCLOUD

```python
# Plotting SoundCloud followers and plays over time
fig, ax1 = plt.subplots(figsize=(12, 7))

color = 'tab:purple'
ax1.set_xlabel('Date')
ax1.set_ylabel('SoundCloud Followers', color=color)
ax1.plot(growth_df['Date'], growth_df['SoundCloud followers'], color=color, label='SoundCloud Followers')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:pink'
ax2.set_ylabel('SoundCloud Plays', color=color)  # we already handled the x-label with ax1
ax2.plot(growth_df['Date'], growth_df['SoundCloud plays'], color=color, label='SoundCloud Plays', linestyle='--')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('SoundCloud Followers and Plays Over Time')
plt.show()
```

## CONNECTIONS AND INTERATCTION

## SOUNDCLOUD ANALISYS

```python
# Loading the three provided CSV files related to Mattia Saviolo's SoundCloud data
# to prepare for an analysis of his SoundCloud presence, including plays, followers growth, and track information.

# File paths for the uploaded CSV files
plays_path = 'Mattia Saviolo-SoundCloud Plays.csv'
followers_growth_path = 'Mattia Saviolo-SoundCloud Followers Growth.csv'
tracks_path = 'Mattia Saviolo-SoundCloud Tracks.csv'

# Loading each file into a Pandas DataFrame
plays_df = pd.read_csv(plays_path)
followers_growth_df = pd.read_csv(followers_growth_path)
tracks_df = pd.read_csv(tracks_path)

# Displaying the first few rows of each DataFrame to understand their structure
plays_df.head(), followers_growth_df.head(), tracks_df.head()
```

```python
# Convert 'Date' columns to datetime for better time series analysis
plays_df['Date'] = pd.to_datetime(plays_df['Date'])
followers_growth_df['Date'] = pd.to_datetime(followers_growth_df['Date'])
tracks_df['Release date'] = pd.to_datetime(tracks_df['Release date'])

# Analysis 1: Trend of Plays over Time
# Aggregating weekly plays data to visualize the trend

# Plotting Plays over Time
plt.figure(figsize=(14, 7))
plt.plot(plays_df['Date'], plays_df['Plays'], marker='o', linestyle='-', color='royalblue')
plt.title('SoundCloud Plays Over Time for Mattia Saviolo')
plt.xlabel('Date')
plt.ylabel('Plays')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Analysis 2: Followers Growth Over Time
# Visualizing the weekly growth in followers to understand how Mattia Saviolo's fanbase on SoundCloud has developed

plt.figure(figsize=(14, 7))
plt.plot(followers_growth_df['Date'], followers_growth_df['Total followers'], marker='o', linestyle='-', color='darkorange')
plt.title('SoundCloud Total Followers Growth Over Time for Mattia Saviolo')
plt.xlabel('Date')
plt.ylabel('Total Followers')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Analysis 3: Popularity of Tracks on SoundCloud
# Identifying the most popular tracks based on all-time plays

# Sorting tracks by 'Alltime plays' to identify the most popular ones
top_tracks_by_plays = tracks_df.sort_values(by='Alltime plays', ascending=False).head(5)

# Plotting the top 5 most popular tracks
plt.figure(figsize=(12, 8))
plt.barh(top_tracks_by_plays['Track'], top_tracks_by_plays['Alltime plays'], color='green')
plt.xlabel('Alltime Plays')
plt.title('Top 5 Most Popular Tracks by Alltime Plays on SoundCloud for Mattia Saviolo')
plt.gca().invert_yaxis()  # To have the track with the most plays on top
plt.tight_layout()
plt.show()
```
