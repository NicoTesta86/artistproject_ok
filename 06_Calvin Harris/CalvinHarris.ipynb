```python
career_performance_path = 'Calvin Harris-Analytics-Career Performance.csv'
career_performance_df = pd.read_csv(career_performance_path)

career_performance_df.head()
```

The "Calvin Harris-Analytics-Career Performance.csv" file tracks the performance of Calvin Harris over time, showing the evolution of his "Viberate performance points" and his rank according to Viberate. Here's a brief overview of the data provided:

- Date: The data begins from February 27, 2023, and continues onwards, providing a timeline of his performance metrics.
    Viberate Performance Points: These points seem to represent a cumulative or overall measure of his performance in the industry, starting at 14,943,800 points and showing changes over the days.
- Viberate Rank: His rank according to Viberate, which appears on February 28, 2023, at 53rd position, and remains constant in the snapshot provided.

This dataset can be used to analyze trends in Calvin Harris's career performance over time, potentially identifying peaks or declines in his popularity or succes

```python
import matplotlib.pyplot as plt

career_performance_df['Date'] = pd.to_datetime(career_performance_df['Date'])

plt.figure(figsize=(12, 6))
plt.plot(career_performance_df['Date'], career_performance_df['Viberate performance points'], marker='o', linestyle='-', color='blue')
plt.title('Calvin Harris Career Performance Over Time')
plt.xlabel('Date')
plt.ylabel('Viberate Performance Points')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout() 

plt.show()
```

The plot above illustrates the trend in Calvin Harris's career performance over time, as measured by Viberate performance points. From the provided data snippet, we observe a general upward trend in his performance points, indicating a positive trajectory in his career metrics during this period.

This visualization helps to quickly grasp the overall performance and could be enriched with a more detailed dataset spanning longer periods or incorporating additional metrics for a comprehensive analysis.

The "Calvin Harris-Analytics-Fanbase Daily Growth and Engagement.csv" file details the daily changes and engagement across Calvin Harris's major social and streaming platforms. This dataset provides a snapshot starting from March 27, 2023, and includes the following metrics:

- Spotify Followers: Total followers and daily change.
- Spotify Monthly Listeners: Total monthly listeners and daily change.
- YouTube Subscribers: Total subscribers and daily change.
- TikTok Followers: Total followers, daily change, likes, and views.
- Instagram Followers: Total followers, daily change, and likes.
- SoundCloud Followers: Total followers, daily change, and plays.

This comprehensive dataset allows for a detailed analysis of the daily growth and engagement of Calvin Harris's fanbase on key digital platforms. By analyzing these metrics, we can identify trends, such as periods of significant growth or engagement spikes, which might be correlated with album releases, single drops, or other significant events in his career.

```python
fanbase_daily_growth_df['Date'] = pd.to_datetime(fanbase_daily_growth_df['Date'])

fig, ax1 = plt.subplots(figsize=(14, 7))

color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Spotify Followers', color=color)
ax1.plot(fanbase_daily_growth_df['Date'], fanbase_daily_growth_df['Spotify followers'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  
color = 'tab:blue'
ax2.set_ylabel('Daily Change in Spotify Followers', color=color)  #
ax2.plot(fanbase_daily_growth_df['Date'], fanbase_daily_growth_df['Spotify followers change'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  
plt.title('Calvin Harris Spotify Followers Growth and Daily Change')
plt.show()
```

The plot illustrates Calvin Harris's Spotify followers' growth alongside the daily change in followers. The red line represents the total number of Spotify followers over time, showing a general upward trend. The blue line indicates the daily change in followers, providing insight into how his fanbase grows day by day.

This visualization helps to understand the dynamics of his Spotify presence, showing not just the overall growth but also the variability in how his fanbase expands on a daily basis. It's interesting to note periods where there might be spikes in daily growth, which could correlate with new music releases, promotional activities, or other events.
