## Gorgon City

To provide a General Overview of Gorgon City's performance on SoundCloud and Spotify, we'll look into several key aspects from the datasets you've provided. This overview will cover:

- Total Followers and Listener Engagement: Comparing the total number of followers on SoundCloud and Spotify to see which platform has a larger audience base for Gorgon City.
- Growth Over Time: Observing the trend in follower growth on both platforms to understand where the audience is expanding more rapidly.
- Streaming Numbers: Evaluating the total streams or plays Gorgon City has received on each platform to gauge overall engagement.
- Track Popularity: Highlighting the most popular tracks on both platforms, based on streams/plays, to see if there are differences in listener preferences across platforms.

Let's start by loading the datasets to extract the necessary information for this overview. We'll begin with the follower growth and overview datasets for both platforms.

```python
import pandas as pd

# Load the datasets
sc_followers_growth = pd.read_csv('Gorgon City-SoundCloud Followers Growth.csv')
sc_overview = pd.read_csv('Gorgon City-SoundCloud Overview.csv')
spotify_followers_growth = pd.read_csv('Gorgon City-Spotify Follower Growth and Popularity.csv')
spotify_overview = pd.read_csv('Gorgon City-Spotify Overview.csv')

# Display the first few rows of each dataset for an initial understanding
datasets = {
    "SoundCloud Followers Growth": sc_followers_growth.head(),
    "SoundCloud Overview": sc_overview.head(),
    "Spotify Follower Growth and Popularity": spotify_followers_growth.head(),
    "Spotify Overview": spotify_overview.head()
}

datasets
```

General Overview of Gorgon City's Performance on SoundCloud and Spotify

Based on the initial look at the datasets, here are some key findings and what we can analyze further:
SoundCloud:

- Followers Growth: We have detailed weekly data on new followers and the total number of followers. For example, in April 2023, there was a consistent increase in followers each week, with the total reaching 308,892 by the end of the month.
- Overall Engagement: The "SoundCloud Overview" data shows monthly metrics including followers, plays, likes, comments, and reposts. For instance, in April 2023, there were 856 new followers, 351,499 plays, and 7,142 likes.

Spotify:

- Follower Growth and Popularity: This dataset provides a monthly snapshot of followers, follower growth, and popularity. For instance, by June 2021, Gorgon City had 359,744 followers, with a growth of 4,834 that month, and a popularity score of 73.
- Overall Performance: The "Spotify Overview" dataset includes monthly data on followers, monthly listeners, monthly streams, and ratios indicating engagement. In March 2021, there were increases in both monthly listeners (to 5,690,049) and streams (3,450,129).

Next Steps for Analysis:

- Trend Analysis: We can plot the growth of followers on both platforms over time to visualize trends and identify any significant changes or patterns.
- Engagement Metrics: Compare the average monthly plays/streams and engagement metrics (likes, comments, reposts) on SoundCloud to monthly listeners and streams on Spotify. This will help us understand where Gorgon City's music is more actively engaged with.
    Platform Performance: Assess which platform shows stronger growth and engagement trends for Gorgon City.

```python
import matplotlib.pyplot as plt

# Preparing data for plotting
# SoundCloud Followers Growth
sc_dates = pd.to_datetime(sc_followers_growth['Date'])
sc_followers = sc_followers_growth['Total followers']

# Spotify Follower Growth
spotify_dates = pd.to_datetime(spotify_followers_growth['Date'])
spotify_followers = spotify_followers_growth['Followers']

# Plotting the follower growth on both platforms
plt.figure(figsize=(14, 7))

plt.plot(sc_dates, sc_followers, label='SoundCloud', color='orange')
plt.plot(spotify_dates, spotify_followers, label='Spotify', color='green')

plt.title('Gorgon City Follower Growth on SoundCloud and Spotify')
plt.xlabel('Date')
plt.ylabel('Total Followers')
plt.legend()
plt.grid(True)

plt.show()
```

The trend graph above illustrates Gorgon City's follower growth on SoundCloud and Spotify over the periods covered in the datasets. Here are some observations:

- Growth Patterns: Both platforms show an upward trend in the number of followers, indicating increasing popularity of Gorgon City on these platforms.
- Platform Comparison: While the actual numbers differ due to the inherent audience sizes and platform differences, both SoundCloud and Spotify show significant growth, suggesting effective engagement and outreach strategies.

Next, we can compare engagement metrics (such as plays/streams, likes, and comments) on SoundCloud to monthly listeners and streams on Spotify to better understand audience engagement on these platforms.

```python
# Calculating average engagement metrics for SoundCloud
sc_avg_plays = sc_overview['Soundcloud plays'].mean()
sc_avg_likes = sc_overview['Soundcloud likes'].mean()
sc_avg_comments = sc_overview['Soundcloud comments'].mean()
sc_avg_reposts = sc_overview['Soundcloud reposts'].mean()

# Calculating average engagement metrics for Spotify
spotify_avg_listeners = spotify_overview['Monthly listeners'].mean()
spotify_avg_streams = spotify_overview['Monthly streams'].mean()

# Preparing data for comparison
sc_metrics = [sc_avg_plays, sc_avg_likes, sc_avg_comments, sc_avg_reposts]
spotify_metrics = [spotify_avg_listeners, spotify_avg_streams, 0, 0]  # Placeholder zeros for likes/comments as not directly comparable
labels = ['Plays/Streams', 'Likes', 'Comments', 'Reposts']

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.35
index = range(len(labels))

bar1 = ax.bar(index, sc_metrics, bar_width, label='SoundCloud', color='orange')
bar2 = ax.bar([p + bar_width for p in index], spotify_metrics, bar_width, label='Spotify', color='green')

ax.set_xlabel('Metric')
ax.set_ylabel('Average Count')
ax.set_title('Average Engagement Metrics on SoundCloud and Spotify')
ax.set_xticks([p + bar_width / 2 for p in index])
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```


The comparison of average engagement metrics on SoundCloud and Spotify reveals several insights into Gorgon City's audience engagement across these platforms:

- Plays/Streams: Both platforms show substantial engagement, with a considerable number of plays/streams. The metrics indicate active listening habits among the audience on both platforms.
- Likes: On SoundCloud, the average number of likes provides a direct measure of listener appreciation, whereas Spotify doesn't provide a direct equivalent for likes in the provided dataset.
- Comments and Reposts: SoundCloud's engagement metrics include comments and reposts, showing additional layers of audience interaction that aren't directly comparable with Spotify's metrics in the provided dataset.
- Monthly Listeners on Spotify: The average monthly listeners on Spotify is a unique metric that illustrates the breadth of Gorgon City's reach on the platform.

This analysis shows strong audience engagement on both platforms, with SoundCloud providing a more interactive set of metrics (likes, comments, reposts) and Spotify showcasing extensive reach through monthly listeners and stream counts. The differences highlight the platforms' unique strengths: SoundCloud in audience interaction and Spotify in broad listener reach.
